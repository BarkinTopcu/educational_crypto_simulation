```python
!pip install modin[ray] pandas
import pandas as pd
import dask.dataframe as dd
```

```python
data_18 = dd.read_csv("btc_201801010000_201901010000.csv")
data_19 = dd.read_csv("btc_201901010000_202001010000.csv")
data_20 = dd.read_csv("btc_202001010000_202101010000.csv")
data_21 = dd.read_csv("btc_202101010000_202201010000.csv")
data_22 = dd.read_csv("btc_202201010000_202301010000.csv")
data_23 = dd.read_csv("btc_202301010000_202401010000.csv")
data_24 = dd.read_csv("btc_202401010000_202501010000.csv")
data_25 = dd.read_csv("btc_2025_min.csv")
```

```python
data_18 = data_18.compute()
data_19 = data_19.compute()
data_20 = data_20.compute()
data_21 = data_21.compute()
data_22 = data_22.compute()
data_23 = data_23.compute()
data_24 = data_24.compute()
data_25 = data_25.compute()
```

```python
data_20.info()
```

```python
data_5mn = pd.concat([data_18,data_19,data_20,data_21,data_22,data_23,data_24,data_25],ignore_index=True)
```

```python
data_5mn
```

```python
data_5mn.drop(columns=["close_time","ignore"],inplace=True)
```

```python
data_5mn.drop_duplicates(inplace=True)
```

```python
data_5mn.info()
```

```python
def get_duplicate_timestamp_counts(df):
    # Eƒüer timestamp string ise, datetime'a √ßevir
    df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Her bir timestamp'in ka√ß kez tekrar ettiƒüini say
    counts = df['timestamp'].value_counts()

    # Sadece birden fazla tekrar edenleri filtrele
    duplicate_counts = counts[counts > 1]

    return duplicate_counts
```

```python
duplicate_counts = get_duplicate_timestamp_counts(data_5mn)

# ƒ∞lk 10 sonucu g√∂rmek i√ßin:
print(duplicate_counts.head(10))
```

```python
data_5mn['timestamp'] = pd.to_datetime(data_5mn['timestamp'])
```

```python
data_5mn.set_index('timestamp', inplace=True)
```

```python
data_5mn
```

```python
result = data_5mn.resample('5T', label='right', closed='right').agg({
    'volume': 'sum',
    'quote_asset_volume': 'sum',
    'num_trades': 'sum',
    'taker_buy_base': 'sum',
    'taker_buy_quote': 'sum'
}).reset_index()
```

```python
result.rename(columns={
    'volume': 'sum_volume_last_5min',
    'quote_asset_volume': 'sum_quote_volume_last_5min',
    'num_trades': 'sum_trades_last_5min',
    'taker_buy_base': 'sum_taker_base_last_5min',
    'taker_buy_quote': 'sum_taker_quote_last_5min'
}, inplace=True)
```

```python
result
```

```python
result['timestamp'] = pd.to_datetime(result['timestamp'])
```

```python
data_5mn = data_5mn.reset_index()
```

```python
data_5mn
```

```python
merged_df = pd.merge(result, data_5mn, on='timestamp', how='left')
```

```python
merged_df
```

```python
merged_df['year'] = merged_df['timestamp'].dt.year
merged_df['month'] = merged_df['timestamp'].dt.month
merged_df['day'] = merged_df['timestamp'].dt.day
merged_df['hour'] = merged_df['timestamp'].dt.hour
merged_df['minute'] = merged_df['timestamp'].dt.minute
```

```python
merged_df
```

```python
merged_df.to_csv("btc_5mn_data.csv",index=False)
```
