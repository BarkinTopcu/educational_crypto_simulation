```python
def batch_backtest(data_sets, backtest_func, fee_rate=0.001, initial_balance=100):
    """
    Ã‡oklu DataFrame ve backtest fonksiyonuyla toplu backtest yapar.

    Args:
        data_sets (list of tuples): [(isim, df), ...] ÅŸeklinde dataframeler.
        backtest_func (function): Uygulanacak backtest fonksiyonu.
        fee_rate (float): Komisyon oranÄ±.
        initial_balance (float): BaÅŸlangÄ±Ã§ bakiyesi.

    Returns:
        pd.DataFrame: SonuÃ§ tablosu.
    """
    results = []
    for name, df in data_sets:
        final, fee, signal_count = backtest_func(df, fee_rate=fee_rate, initial_balance=initial_balance)
        results.append({
            'DÃ¶nem': name,
            'BaÅŸlangÄ±Ã§ Bakiye': initial_balance,
            'Final Bakiye': final,
            'Toplam Fee': fee,
            'Toplam Ä°ÅŸlem SayÄ±sÄ±': signal_count
        })
    results = pd.DataFrame(results)
    return results
```

```python
def sma_crossover_backtest(df, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)
    df['SMA_5'] = df['close'].rolling(window=5).mean()
    df['SMA_20'] = df['close'].rolling(window=20).mean()
    df['signal'] = 0
    df.loc[(df['SMA_5'] > df['SMA_20']) & (df['SMA_5'].shift(1) <= df['SMA_20'].shift(1)), 'signal'] = 1
    df.loc[(df['SMA_5'] < df['SMA_20']) & (df['SMA_5'].shift(1) >= df['SMA_20'].shift(1)), 'signal'] = -1

    balance = initial_balance
    fee_paid = 0.0
    position = 0
    entry_price = 0.0
    coin_qty = 0.0

    equity_list, position_list, signal_list, fee_list = [], [], [], []

    for i, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            entry_price = price
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)
        signal_list.append(signal)
        position_list.append(position)
        fee_list.append(fee_paid)

    df['sim_signal'] = signal_list
    df['position'] = position_list
    df['equity'] = equity_list
    df['fee_total'] = fee_list


    signal_count = sum(1 for s in signal_list if s != 0)
    return equity_list[-1], fee_paid, signal_count
```

```python
# 1. Parametrik SMA backtest fonksiyonu
def sma_crossover_backtest_parametric(df, fast_sma, slow_sma, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)
    df['SMA_fast'] = df['close'].rolling(window=fast_sma).mean()
    df['SMA_slow'] = df['close'].rolling(window=slow_sma).mean()

    df['signal'] = 0
    df.loc[(df['SMA_fast'] > df['SMA_slow']) & (df['SMA_fast'].shift(1) <= df['SMA_slow'].shift(1)), 'signal'] = 1
    df.loc[(df['SMA_fast'] < df['SMA_slow']) & (df['SMA_fast'].shift(1) >= df['SMA_slow'].shift(1)), 'signal'] = -1

    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * 0.001
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * 0.001
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    total_trades = df['signal'].abs().sum()

    return final_equity, fee_paid, total_trades


# 2. Grid search + heatmap Ã§izen fonksiyon
def grid_search_with_heatmap(datasets, fast_range=range(3, 15), slow_range=range(10, 60, 5)):
    for name, df in datasets:
        result = []
        for fast in fast_range:
            for slow in slow_range:
                if fast >= slow:
                    continue
                final_balance, fee_paid, trade_count = sma_crossover_backtest_parametric(df, fast, slow)
                result.append({
                    'Fast': fast,
                    'Slow': slow,
                    'FinalBalance': final_balance
                })

        result_df = pd.DataFrame(result)
        pivot = result_df.pivot(index='Fast', columns='Slow', values='FinalBalance')

        plt.figure(figsize=(10, 8))
        sns.heatmap(pivot, annot=True, fmt=".2f", cmap='YlGnBu')
        plt.title(f"{name} - SMA Grid Search Final Balance Heatmap")
        plt.xlabel("Slow SMA")
        plt.ylabel("Fast SMA")
        plt.tight_layout()
        plt.show()

        # En iyi sonucu yazdÄ±r
        best_row = result_df.loc[result_df['FinalBalance'].idxmax()]
        print(f"ðŸ“Š {name} â†’ En iyi SMA: Fast={int(best_row['Fast'])}, Slow={int(best_row['Slow'])}, Final Balance: {best_row['FinalBalance']:.2f}")
```

```python
def rsi_backtest(df, rsi_period=14, overbought=70, oversold=30, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)

    # RSI hesapla
    df['RSI'] = ta.momentum.RSIIndicator(df['close'], window=rsi_period).rsi()
    df['signal'] = 0

    # Al: RSI oversold'dan yukarÄ± keserse
    df.loc[(df['RSI'] < oversold) & (df['RSI'].shift(1) >= oversold), 'signal'] = 1

    # Sat: RSI overbought'dan aÅŸaÄŸÄ± keserse
    df.loc[(df['RSI'] > overbought) & (df['RSI'].shift(1) <= overbought), 'signal'] = -1

    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    total_trades = df['signal'].abs().sum()

    return final_equity, fee_paid, total_trades
def rsi_grid_search_with_heatmap(datasets, rsi_range=range(7, 22, 2), oversold_range=range(20, 35, 5), overbought_range=range(65, 85, 5)):
    for name, df in datasets:
        results = []

        for rsi_period in rsi_range:
            for oversold in oversold_range:
                for overbought in overbought_range:
                    if oversold >= overbought:
                        continue
                    final_balance, fee, trades = rsi_backtest(df, rsi_period, overbought, oversold)
                    results.append({
                        'RSI_Period': rsi_period,
                        'Overbought': overbought,
                        'Oversold': oversold,
                        'FinalBalance': final_balance
                    })

        result_df = pd.DataFrame(results)

        # SeÃ§: en iyi RSI periyodu iÃ§in oversold/overbought heatmap
        best_period = result_df.groupby('RSI_Period')['FinalBalance'].mean().idxmax()
        best_subset = result_df[result_df['RSI_Period'] == best_period]

        pivot = best_subset.pivot(index='Oversold', columns='Overbought', values='FinalBalance')

        plt.figure(figsize=(10, 8))
        sns.heatmap(pivot, annot=True, fmt=".2f", cmap='YlOrRd')
        plt.title(f"{name} - RSI Grid Search (RSI={best_period})")
        plt.xlabel("Overbought")
        plt.ylabel("Oversold")
        plt.tight_layout()
        plt.show()

        # En iyi sonucu yazdÄ±r
        best_row = result_df.loc[result_df['FinalBalance'].idxmax()]
        print(f"ðŸ“Š {name} â†’ En iyi RSI: period={int(best_row['RSI_Period'])}, overbought={int(best_row['Overbought'])}, oversold={int(best_row['Oversold'])}, Final Balance: {best_row['FinalBalance']:.2f}")
```

```python
def qvd_backtest(df, qvd_threshold=1.5, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)

    # QVD hesapla
    df['QVD'] = df['quote_asset_volume'] / (df['volume'] + 1e-9)  # sÄ±fÄ±ra bÃ¶lme hatasÄ±nÄ± engeller

    # Fiyat momentumu iÃ§in referans
    df['price_change'] = df['close'].diff()

    df['signal'] = 0

    # Al sinyali: QVD yÃ¼kseliyor ve fiyat da artÄ±yor â†’ kurumsal destekli
    df.loc[(df['QVD'] > qvd_threshold) & (df['price_change'] > 0), 'signal'] = 1

    # Sat sinyali: QVD dÃ¼ÅŸÃ¼yor ama fiyat hÃ¢lÃ¢ yÃ¼kseliyor â†’ zayÄ±f hareket
    df.loc[(df['QVD'] < 1.0) & (df['price_change'] < 0), 'signal'] = -1

    # Backtest simÃ¼lasyonu
    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    total_trades = df['signal'].abs().sum()

    return final_equity, fee_paid, total_trades
```

```python
def qvd_grid_search_with_heatmap(datasets, qvd_range=[1.0, 1.2, 1.5, 1.8, 2.0]):
    for name, df in datasets:
        results = []

        for threshold in qvd_range:
            final, fee, trades = qvd_backtest(df, qvd_threshold=threshold)
            results.append({
                'Threshold': threshold,
                'FinalBalance': final
            })

        result_df = pd.DataFrame(results)
        pivot = result_df.pivot_table(index='Threshold', values='FinalBalance')

        plt.figure(figsize=(8, 6))
        sns.heatmap(pivot, annot=True, fmt=".2f", cmap='coolwarm')
        plt.title(f"{name} - Quote Volume Divergence Heatmap")
        plt.xlabel("QVD Threshold")
        plt.ylabel("")
        plt.tight_layout()
        plt.show()

        # En iyi sonucu yazdÄ±r
        best_row = result_df.loc[result_df['FinalBalance'].idxmax()]
        print(f"ðŸ“Š {name} â†’ En iyi QVD Threshold = {best_row['Threshold']} â†’ Final Balance = {best_row['FinalBalance']:.2f}")
```

```python
def vwap_reversion_backtest(df, threshold=0.01, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)

    # VWAP hesapla
    df['vwap_numerator'] = (df['close'] * df['volume']).cumsum()
    df['vwap_denominator'] = df['volume'].cumsum() + 1e-9  # sÄ±fÄ±ra bÃ¶lme engeli
    df['VWAP'] = df['vwap_numerator'] / df['vwap_denominator']

    df['deviation'] = (df['close'] - df['VWAP']) / df['VWAP']

    df['signal'] = 0

    # AL: fiyat VWAP'in altÄ±na %threshold kadar inmiÅŸse
    df.loc[df['deviation'] < -threshold, 'signal'] = 1

    # SAT: fiyat VWAP'in Ã¼stÃ¼ne %threshold kadar Ã§Ä±kmÄ±ÅŸsa
    df.loc[df['deviation'] > threshold, 'signal'] = -1

    # Backtest simÃ¼lasyonu
    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    trade_count = df['signal'].abs().sum()

    return final_equity, fee_paid, trade_count
```

```python
def vwap_grid_search_with_heatmap(datasets, threshold_range=[0.005, 0.01, 0.015, 0.02, 0.025]):
    for name, df in datasets:
        results = []

        for threshold in threshold_range:
            final, fee, trades = vwap_reversion_backtest(df, threshold=threshold)
            results.append({
                'Threshold': threshold,
                'FinalBalance': final
            })

        result_df = pd.DataFrame(results)
        pivot = result_df.pivot_table(index='Threshold', values='FinalBalance')

        plt.figure(figsize=(8, 6))
        sns.heatmap(pivot, annot=True, fmt=".2f", cmap='Spectral')
        plt.title(f"{name} - VWAP Reversion Heatmap")
        plt.xlabel("VWAP Deviation Threshold")
        plt.tight_layout()
        plt.show()

        best_row = result_df.loc[result_df['FinalBalance'].idxmax()]
        print(f"ðŸ“Š {name} â†’ En iyi threshold = {best_row['Threshold']} â†’ Final: {best_row['FinalBalance']:.2f}")
```

```python
def doji_reversal_backtest(df, body_threshold=0.001, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)

    # Mum gÃ¶vdesi oranÄ±: |close - open| / (high - low)
    df['body_size'] = abs(df['close'] - df['open'])
    df['range'] = df['high'] - df['low'] + 1e-9  # sÄ±fÄ±ra bÃ¶lme korumasÄ±
    df['body_ratio'] = df['body_size'] / df['range']

    df['signal'] = 0

    # Doji tespiti: gÃ¶vde Ã§ok kÃ¼Ã§Ã¼k
    df['is_doji'] = df['body_ratio'] < body_threshold

    # AL sinyali: doji â†’ ardÄ±ndan yukarÄ± mum (bullish reversal)
    df['next_close'] = df['close'].shift(-1)
    df['next_open'] = df['open'].shift(-1)
    df.loc[(df['is_doji']) & (df['next_close'] > df['next_open']), 'signal'] = 1

    # SAT sinyali: doji â†’ ardÄ±ndan aÅŸaÄŸÄ± mum (bearish reversal)
    df.loc[(df['is_doji']) & (df['next_close'] < df['next_open']), 'signal'] = -1

    # Backtest simÃ¼lasyonu
    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    total_trades = df['signal'].abs().sum()

    return final_equity, fee_paid, total_trades
```

```python
def doji_grid_search_with_heatmap(datasets, threshold_range=[0.005, 0.01, 0.02, 0.03, 0.05, 0.1]):
    for name, df in datasets:
        results = []
        for threshold in threshold_range:
            final, fee, trades = doji_reversal_backtest(df, body_threshold=threshold)
            results.append({
                'Body_Threshold': threshold,
                'FinalBalance': final
            })

        result_df = pd.DataFrame(results)
        pivot = result_df.pivot_table(index='Body_Threshold', values='FinalBalance')

        plt.figure(figsize=(8, 6))
        sns.heatmap(pivot, annot=True, fmt=".2f", cmap='YlGnBu')
        plt.title(f"{name} - Doji Reversal Heatmap")
        plt.xlabel("Body Size Threshold")
        plt.tight_layout()
        plt.show()

        best = result_df.loc[result_df['FinalBalance'].idxmax()]
        print(f"ðŸ“Š {name} â†’ En iyi threshold: {best['Body_Threshold']} â†’ Final: {best['FinalBalance']:.2f}")
```

```python
def buy_pressure_backtest(df, bpd_threshold=0.6, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)

    # Buy Pressure Dominance hesapla
    df['BPD'] = df['taker_buy_base'] / (df['volume'] + 1e-9)

    df['signal'] = 0

    # AL sinyali: BPD yÃ¼ksek
    df.loc[(df['BPD'] > bpd_threshold) & (df['BPD'].shift(1) <= bpd_threshold), 'signal'] = 1

    # SAT sinyali: BPD dÃ¼ÅŸÃ¼yor
    df.loc[(df['BPD'] < bpd_threshold) & (df['BPD'].shift(1) >= bpd_threshold), 'signal'] = -1

    # Backtest simÃ¼lasyonu
    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    total_trades = df['signal'].abs().sum()

    return final_equity, fee_paid, total_trades
```

```python
def bullish_engulfing_backtest(df, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)

    df['signal'] = 0

    # Engulfing tanÄ±mÄ±
    prev_open = df['open'].shift(1)
    prev_close = df['close'].shift(1)
    curr_open = df['open']
    curr_close = df['close']

    # Ã–nceki mum kÄ±rmÄ±zÄ±, bu mum yeÅŸil ve bu mumun gÃ¶vdesi daha bÃ¼yÃ¼k
    df.loc[
        (prev_close < prev_open) &
        (curr_close > curr_open) &
        (curr_close > prev_open) &
        (curr_open < prev_close),
        'signal'
    ] = 1  # AL sinyali

    # SatÄ±ÅŸ sinyali: fiyat dÃ¼ÅŸmeye baÅŸlarsa (Ã¶rnek: 2 kÄ±rmÄ±zÄ± mum)
    df['down1'] = df['close'] < df['close'].shift(1)
    df['down2'] = df['close'].shift(1) < df['close'].shift(2)
    df.loc[df['down1'] & df['down2'], 'signal'] = -1

    # Backtest motoru
    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    trade_count = df['signal'].abs().sum()

    return final_equity, fee_paid, trade_count
```

```python
def doji_trend_backtest(df, body_threshold=0.05, sma_period=20, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)

    # GÃ¶vde oranÄ±
    df['body_size'] = abs(df['close'] - df['open'])
    df['range'] = df['high'] - df['low'] + 1e-9  # sÄ±fÄ±ra bÃ¶lme korumasÄ±
    df['body_ratio'] = df['body_size'] / df['range']

    # SMA (trend filtresi)
    df['SMA'] = df['close'].rolling(sma_period).mean()

    df['signal'] = 0

    # Doji + Trend dÃ¶nÃ¼ÅŸÃ¼ = Al
    df['next_open'] = df['open'].shift(-1)
    df['next_close'] = df['close'].shift(-1)

    df.loc[
        (df['body_ratio'] < body_threshold) &
        (df['next_close'] > df['next_open']) &
        (df['next_close'] > df['SMA']),
        'signal'
    ] = 1

    # Ã‡Ä±kÄ±ÅŸ: fiyat SMA'nÄ±n altÄ±na dÃ¼ÅŸerse
    df.loc[(df['close'] < df['SMA']) & (df['signal'] == 0), 'signal'] = -1

    # Backtest motoru
    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    total_trades = df['signal'].abs().sum()

    return final_equity, fee_paid, total_trades
```

```python
def engulfing_rsi_backtest(df, rsi_threshold=30, rsi_period=14, fee_rate=0.001, initial_balance=100):
    df = df.copy().reset_index(drop=True)

    # RSI
    df['RSI'] = ta.momentum.RSIIndicator(df['close'], window=rsi_period).rsi()

    # Engulfing
    prev_open = df['open'].shift(1)
    prev_close = df['close'].shift(1)
    curr_open = df['open']
    curr_close = df['close']
    bullish = (prev_close < prev_open) & (curr_close > curr_open) & (curr_close > prev_open) & (curr_open < prev_close)

    df['signal'] = 0
    df.loc[bullish & (df['RSI'] < rsi_threshold), 'signal'] = 1
    df.loc[df['RSI'] > 70, 'signal'] = -1

    # Backtest
    balance = initial_balance
    fee_paid = 0.0
    position = 0
    coin_qty = 0.0
    equity_list = []

    for _, row in df.iterrows():
        signal = row['signal']
        price = row['close']

        if signal == 1 and position == 0:
            fee = balance * fee_rate
            coin_qty = (balance - fee) / price
            fee_paid += fee
            balance = 0.0
            position = 1

        elif signal == -1 and position == 1:
            gross_balance = coin_qty * price
            fee = gross_balance * fee_rate
            balance = gross_balance - fee
            fee_paid += fee
            coin_qty = 0.0
            position = 0

        equity = balance if position == 0 else coin_qty * price
        equity_list.append(equity)

    final_equity = equity_list[-1] if equity_list else initial_balance
    trade_count = df['signal'].abs().sum()

    return final_equity, fee_paid, trade_count
```

```python
def engulfing_rsi_grid_search_with_heatmap(datasets, rsi_range=range(20, 41, 5), rsi_periods=[10, 14, 20]):
    for name, df in datasets:
        results = []

        for rsi_th in rsi_range:
            for rsi_p in rsi_periods:
                final, fee, trades = engulfing_rsi_backtest(df, rsi_threshold=rsi_th, rsi_period=rsi_p)
                results.append({
                    'RSI_Threshold': rsi_th,
                    'RSI_Period': rsi_p,
                    'FinalBalance': final
                })

        result_df = pd.DataFrame(results)
        pivot = result_df.pivot(index='RSI_Threshold', columns='RSI_Period', values='FinalBalance')

        plt.figure(figsize=(10, 7))
        sns.heatmap(pivot, annot=True, fmt=".2f", cmap='crest')
        plt.title(f"{name} - Engulfing + RSI Filter")
        plt.xlabel("RSI Period")
        plt.ylabel("RSI Threshold")
        plt.tight_layout()
        plt.show()

        best = result_df.loc[result_df['FinalBalance'].idxmax()]
        print(f"ðŸ“Š {name} â†’ En iyi: RSI < {best['RSI_Threshold']} | Period={best['RSI_Period']} â†’ Final: {best['FinalBalance']:.2f}")
```
